```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
calls = pd.read_csv("oct1_oct2_2023_311_Service_Requests_from_2010_to_Present.csv", parse_dates =["Created Date"])
```

```python
calls.head(3)
```

```python
sun_filter = calls["Created Date"].dt.day == 1
```

```python
sun_calls = calls[sun_filter]
```

```python
sun_calls.head(2)
```

```python
mon_filter = calls["Created Date"].dt.day == 2
mon_calls = calls[mon_filter]
mon_calls.head(3)
```

```python
sun_calls["Created Date"].dt.hour
```

```python
sun_calls["Created Date"].dt.hour.hist(bins = 30)
```

```python
mon_calls["Created Date"].dt.hour.hist(bins = 30)
```

```python
sun_calls["Created Date"].dt.hour.hist(bins = 30, alpha = 0.5, color = "Blue", density = "True")
mon_calls["Created Date"].dt.hour.hist(bins = 30,alpha = 0.5, color = "red", density = "True")
plt.xlabel("hours")
plt.title( "Total Distribition of complaints")
plt.ylabel("number of complaints")
plt.legend(["Sunday", "Monday"])
```

```python
sun_borough_counts = sun_calls["Borough"].value_counts()
sun_borough_counts
```

```python
mon_borough_counts = mon_calls["Borough"].value_counts()
mon_borough_counts 
```

```python
borough_counts_df = pd.DataFrame({"sunday":sun_borough_counts,"monday":mon_borough_counts})
borough_counts_df
```

```python
borough_counts_df.plot(kind = "bar")
```

```python
borough_counts_df.plot(kind = "bar", color = ["purple","turquoise"])
```

```python
borough_counts_df.plot(kind = "bar", color = ["purple","turquoise"])
plt.xlabel("borough")
plt.title( "Total Distribition of complaints")
plt.ylabel("number of complaints")
plt.legend(["Sunday", "Monday"])
```

```python
before_6am_filter = calls["Created Date"].dt.hour < 6
after_6pm_filter = calls["Created Date"].dt.hour > 18

night_calls = calls[before_6am_filter | after_6pm_filter]
day_calls = calls[~before_6am_filter & ~after_6pm_filter]

night_counts = night_calls["Complaint Type"].value_counts()
day_counts = day_calls["Complaint Type"].value_counts()
night_day_df = pd.DataFrame({"Day":day_counts, "Night": night_counts})

# Plot only the complaints that occur more than 100 times during the day 
# and more than 50 times during the night.
day_top_filter = night_day_df["Day"] > 100
night_top_filter =  night_day_df["Night"] >50
night_day_df[day_top_filter & night_top_filter].plot(kind = "bar")
```
