```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
art = pd.read_csv("Artworks.csv",na_values = ["Unknown", "n.d.", "nd", "n.d", "TBC","TBD"])
art
```

```python
art = art.dropna(subset = ["Gender", "Nationality","Date"])
art.head(2)
```

```python
art["Gender"].value_counts()
```

```python
def count_artists(x):
    return x.count("(")
```

```python
count_artists("(Female)")
```

```python
count_artists("(Male) (Male)")
```

```python
count_artists("() (Male)")
```

```python
art["NumArtists"] = art["Gender"].apply(count_artists)
```

```python
art["NumArtists"]
```

```python
count_artists = art["NumArtists"].value_counts()
count_artists.plot(kind = "bar")
```

```python
def count_americans(x):
    return x.count("American)")
```

```python
art["NumAmericans"] = art["Nationality"].apply(count_americans)
```

```python
art.head()
```

```python
count_americans = art["NumAmericans"].value_counts()
count_americans.plot(kind = "bar")
```
