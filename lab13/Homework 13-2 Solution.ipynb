Are more long green taxi trips taken between midnight and noon than between noon and midnight?

Use you green taxi dataset from Lab 4 (and later labs).  Compute the median trip distance for taxi trips started in the am (between midnight and noon), and the median trip distance for taxi trips started in the pm (between noon and midnight).

Use the `lpep_pickup_datatime` column as the starting time for the trip (so specify it is a date/time column when you read in the CSV file).

```python
import matplotlib.pyplot as plt
import pandas as pd

taxi = pd.read_csv("../../data/Sept1_2021_Green_Taxi_Trip_Data.csv",parse_dates = ["lpep_pickup_datetime"])
taxi.head()
```

```python
# create a dataframe of trips started between midnight and 11:59am
am_filter = taxi["lpep_pickup_datetime"].dt.hour < 12
am_trips = taxi[am_filter]
am_trips.tail()
```

```python
# create a dataframe of trips started between noon and 11:59pm.
pm_filter = taxi["lpep_pickup_datetime"].dt.hour >= 12
pm_trips = taxi[pm_filter]
pm_trips.head()
```

```python
# compute the median am trip distance
am_trips["trip_distance"].median()
```

```python
# compute the median pm trip distance
pm_trips["trip_distance"].median()
```
