Write a piece of code that calculates and displays the empirical distribution of the sample mean.  To do this, write code to take multiple samples from this Met coin dataset, computes the mean of the object begin date for each sample, and displays a histogram of these means.  Use a sample size of 40 and take 800 samples.

```python
import matplotlib.pyplot as plt
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
coins = pd.read_csv("../../../data/MetCoins.csv")
coins.head()
```

```python
means = []
for i in range(800):
    sample = coins.sample(40)
    means.append(sample["Object Begin Date"].mean())
pd.Series(means).hist(bins = 20)
```
