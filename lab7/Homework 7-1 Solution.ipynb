We will use the CSV of coins at the Met Museum from Homework 4-1.  The begin date of an object is the earliest year it could have been created. 

To display all columns use the code `pd.set_option('display.max_columns', None)`.

Assume our population is all the coins in the CSV file.  Write a piece of code that takes 4 random samples of sizes 20, 50, 200, and 1000 from this population and plots the histograms of the object begin dates in each sample.  That is, for each of the 4 sample sizes, take one sample and plot the histogram of its object begin dates.

```python
import matplotlib.pyplot
import pandas as pd
%matplotlib inline

pd.set_option('display.max_columns', None)

coins= pd.read_csv("../../../data/MetCoins.csv")
coins.head()
```

```python
sample20 = coins.sample(20)
sample20["Object Begin Date"].hist(bins = 20)
```

```python
sample50 = coins.sample(50)
sample50["Object Begin Date"].hist(bins = 20)
```

```python
sample200 = coins.sample(200)
sample200["Object Begin Date"].hist(bins = 20)
```

```python
sample1000 = coins.sample(1000)
sample1000["Object Begin Date"].hist(bins = 20)
```
