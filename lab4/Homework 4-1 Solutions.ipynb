The `Object Begin Date` column stores the earliest year the object could have been made, or the year production started.  Write a piece of code that plots a histogram of the Object Begin Date of the coins in the Met's collection.  

Label the x and y axes of your plot and add a title.

```python
import matplotlib.pyplot as plt
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
coins = pd.read_csv("../../data/MetCoins.csv")
```

```python
coins.head()
```

```python
coins["Object Begin Date"].hist(bins=20)
plt.xlabel("Year")
plt.ylabel("# of coins")
plt.title("Begin Date of coins in the Met's collection")
```
