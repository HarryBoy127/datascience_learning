We will use the 311 service request dataset from Lab 8, which contains information about all complaints and service request calls made to 311 in New York city.

Write a piece of code that computes the probability that a 311 call involves a club, bar, or restaurant.  Such calls have `Club/Bar/Restaurant` in the `Location Type` column.

```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
calls = pd.read_csv("../../data/Sept15_2023_311_Service_Requests.csv")
calls.head()
```

```python
# Get the number of calls about a club, bar, or restaurant
location_filter = calls["Location Type"] == "Club/Bar/Restaurant"
num_club_bar_restaurant_calls = location_filter.sum()
num_club_bar_restaurant_calls
```

```python
# Get the total number of complaints.
num_calls = calls.shape[0]
num_calls
```

```python
# Get the probability that a call is about a club, bar, or restaurant
num_club_bar_restaurant_calls/num_calls
```
