## Assignment 24-2:  Passenger fares

This assignment uses the Titanic training data from Lab 24.  You can find information about the different columns in the dataset [here](https://www.kaggle.com/c/titanic/data) (scroll down in the Data Description box at the top).

On the same chart, plot a histogram of the fares paid by the survivors and a histogram of the fares paid by the people who did not survive.  Your histograms should use density instead of frequency on the y axis, and include axis labels, a title, and a legend.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf
%matplotlib inline

train = pd.read_csv("../Data/train.csv")
train.head()
```

```python
survived_filter = train["Survived"] == 1
survived = train[survived_filter]
survived.head()
```

```python
perished_filter = train["Survived"] == 0
perished = train[perished_filter]
perished.head()
```

```python
survived["Fare"].hist(density = True,alpha = 0.5)
perished["Fare"].hist(density = True,alpha = 0.5)
plt.title("Fares paid for voyage on the Titanic")
plt.xlabel("Fare")
plt.ylabel("Density")
plt.legend(["Survivors","Perished"])
```
