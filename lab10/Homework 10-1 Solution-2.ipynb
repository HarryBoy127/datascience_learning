We will use the 311 service request dataset from Lab 8 and 10, which contains information about all complaints and service request calls made to 311 in New York city.

Write a piece of code that computes the probability that a call is from the Bronx or Brooklyn. 

```python
import matplotlib.pyplot as plt
import pandas as pd

calls = pd.read_csv("../../data/Sept15_2023_311_Service_Requests.csv")
pd.set_option('display.max_columns', None)
calls.head(30)
```

```python
# Compute the number of calls from Manhattan or Staten Island.
bronx_filter = calls["Borough"] == "BRONX"
brooklyn_filter = calls["Borough"] == "BROOKLYN"

# Create a dataset of only calls from Manhattan or Staten Island
bronx_or_brooklyn_calls = calls[bronx_filter | brooklyn_filter]
bronx_or_brooklyn_calls.head()
```

```python
num_bronx_or_brooklyn_calls = bronx_or_brooklyn_calls.shape[0]
```

```python
# Compute the total number of calls
num_calls = calls.shape[0]
num_calls
```

```python
# Divide the number of calls from the Manhattan or Staten Island by the total number of calls
# to get the probability that a call is from Manhattan or Staten Island.
num_bronx_or_brooklyn_calls/num_calls
```
